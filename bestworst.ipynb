```python
#Importing the libraries
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import calendar
from datetime import datetime
import yfinance as yf


```

```python
#Create daily list
Mon = []
Tues = []
Wed = []
Thru = []
Fri = []
Sat = []
Sun = []

#create a function to get daily simple return for each day and append it to corres. date
def get_dsr(day, dsr):
    if day == 'Monday':
        Mon.append(float(dsr))
    elif day == 'Tuesday':
        Tues.append(float(dsr))
    elif day == 'Wednesday':
        Wed.append(float(dsr))
    elif day == 'Thrusday':
        Thru.append(float(dsr))
    elif day == 'Friday':
        Fri.append(float(dsr))
    elif day == 'Saturday':
        Sat.append(float(dsr))
    elif day == 'Sunday':
        Sun.append(float(dsr))
    else: 
        print('Something is wrong')
    
```

```python
today = datetime.today().strftime('%Y-%m-%d')
today
```

```python
#Importing the data
df = yf.download('^CRSLDX', start= '2015-01-01', end= today)
#df = df.set_index(pd.DatetimeIndex(df['Date'].values))
#df = pd.read_csv('HDFCBANK.NS.csv')
```

```python
#Calculate and show daily simple return
DSR = df['Close'].pct_change(1)#compute the dsr
df['DSR'] = DSR

df = df[0:]
```

```python
df
```

```python
#Create a loop to gather the daily simple returns of each day
for i in range(0, len(df)):
    df_date = str(df.index[i])#get current date & casting it as a string
    df_dsr = df['DSR'][i]#get current daily simple return
    df_month = df_date.split('-')[0]#get current month
    curr_date = df.index[i]#get the current date
    df_weekday = calendar.day_name[curr_date.weekday()]#get the weekday

    #Add the daily simple returns to the corresponding list for the day
    get_dsr(df_weekday, df_dsr)
```

```python
#Create a function to average the returns of each weekday
def AVG(day):
    if not day:#if this true then list is empty so return a list containing 0
        return[0]
    else:
        return[sum(day)/len(day)]    
```

```python
#Create a df and set the index to be days of week
df_returns = pd.DataFrame(index=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
#Get the average for each weekday and add the values under column AVG
df_returns['AVG'] = AVG(Mon) + AVG(Tues) + AVG(Wed) + AVG(Thru) + AVG(Fri) + AVG(Sat) + AVG(Sun)
```

```python
#Plot the AVG daily simple returns
df_returns.plot.bar()
```
