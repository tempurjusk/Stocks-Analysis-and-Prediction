```python
stockStartDate = '2015-01-01'
today = datetime.today().strftime('%Y-%m-%d')
```

```python
#Importing the data
df = pd.DataFrame()

for stock in Nifty50:
    df[stock] = yf.download(stock, data_source=' yahoo', start= stockStartDate, end= today)['Adj Close']
```

```python
assets = df.columns
```

```python
#Calculate expected annualised returns & sample covariance matrix
mu = expected_returns.mean_historical_return(df)
S = risk_models.sample_cov(df)
```

```python
#Optimize for max Sharp ratio
ef = EfficientFrontier(mu, S) #Create ef object
weights = ef.max_sharpe()

cleaned_weights = ef.clean_weights()
#print(cleaned_weights)
ef.portfolio_performance(verbose=True)
```
