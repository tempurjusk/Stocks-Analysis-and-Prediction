```python
stockStartDate = '2015-01-01'
today = datetime.today().strftime('%Y-%m-%d')
```

```python
Nifty = yf.download('^NSEI', start= stockStartDate, end= today)
Nifty = Nifty.filter(['Close'])
Nifty
```

```python
#Importing the data
df = pd.DataFrame()

for stock in Nifty50:
    df[stock] = yf.download(stock, data_source=' yahoo', start= stockStartDate, end= today)['Adj Close']
```

```python
df
```

```python
df.dropna(axis=1, inplace=True)
df
```

```python
#Get sdr
daily_simple_returns = df.pct_change()
daily_simple_returns
```

```python
dsr_nifty = Nifty.pct_change()
dsr_nifty
```

```python
#Get annualised returns
annual_returns = daily_simple_returns.mean() * 252
annual_returns
```

```python
ar_nifty = dsr_nifty.mean() * 252
ar_nifty
```
