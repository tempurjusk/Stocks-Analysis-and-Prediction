```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import yfinance as yf


```

```python
Year = []
Yearly_Income = []
Yearly_Expense = []
Annual_Returns = []
Yearly_Investments = []
```

```python
income = 8
expense = income/2
interest_rate = 0.08
increment_rate = 0.10
inflation_rate = 0.05
investment = income/2
annual_return = investment + interest_rate
year = 2022
```

```python
Year.append(year)
Yearly_Income.append(income)
Yearly_Expense.append(expense)
Yearly_Investments.append(investment)
Annual_Returns.append(annual_return)
```

```python
#Loop for n years
invested_years = 20
for i in range(0, invested_years-1):
    #Update the investment to be prev investment plus prev annual return plus half income
    investment = investment + annual_return + income/2
    #Update annual returns to be current investment * interest rate
    annual_return = investment* interest_rate
    income = income + (income * increment_rate)
    expense = expense + (expense * inflation_rate)
    #Append new data to lists
    Year.append(year+i+1)
    Yearly_Income.append(income)
    Yearly_Expense.append(expense)
    Yearly_Investments.append(investment)
    Annual_Returns.append(annual_return)
```

```python
#Put all data in a dataframe
df = pd.DataFrame()
df['Year'] = Year
df['Yearly_Income'] = Yearly_Income
df['Yearly_Expense'] = Yearly_Expense
df['Yearly_Investments'] = Yearly_Investments
df['Annual_Returns'] = Annual_Returns
```

```python
#Visualise
plt.figure(figsize=(12.4, 4.5))
plt.plot(df['Year'], df['Yearly_Expense'], label = 'Yearly Expense')
plt.plot(df['Year'], df['Annual_Returns'], label = 'Annual Returns')
plt.xticks(rotation=45)
plt.title('Retirement Plan')
plt.xlabel('Years')
plt.ylabel('Rupees in Lakhs')
plt.xticks(df['Year'])
plt.legend()
plt.show()
```
